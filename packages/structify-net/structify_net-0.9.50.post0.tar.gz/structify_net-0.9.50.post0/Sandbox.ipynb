```python
import numpy as np
import math
```

```python
def proba_function(n,m,epsilon):
    nb_pairs=int(n*(n-1)/2)
    probas_zero=[1]*m+[0]*(nb_pairs-m)
    probas_one=[m/nb_pairs]*nb_pairs
    cum_zero=np.cumsum(probas_zero)
    cum_one=np.cumsum(probas_one)
    #print(cumul_proba_zero)
    #print(cumul_proba_one)
    #zero_component = [(x*(1-epsilon))**2 for x in cumul_proba_zero]
    #one_component = [(x*(epsilon))**2 for x in cumul_proba_one]
    #bary=[math.sqrt(zero_component[i]+one_component[i]) for i in range(len(one_component))]
    bary = [math.log(np.exp(cum_zero[i]-cum_one[i])*epsilon)+cum_one[i] for i in range(len(cum_one))]
    return(bary)
```

```python
n=100

a = proba_function(n,500,0.5)
```

```python
import seaborn as sns
```

```python
sns.lineplot(x=range(int(n*(n-1)/2)),y=a)
```
